# 네이버 csv 정리

제목 / 작가 / 평점 / 연령정보(성인orNot) / 장르(g1 - g12) / 키워드  형태로

## 1. 기존의 크롤링 csv에서 그림작가는 제외한 형태로 가공 (crawling_result_naver_clear.csv)

```
with open('crawling_result_naver.csv', encoding="utf-8") as csv_file:
  with open('crawling_result_naver_clear.csv',mode='w',newline='',encoding='utf-8') as my_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    my_writer = csv.writer(my_file,delimiter=',')
    my_writer.writerow(['idx','name','score','writer','age_limit'])   #,'g1','g2','g3','g4','g5','g6','g7','g8','g9','g10','g11','g12', 'keywords'])

    line_count = 0
    for row in csv_reader:  #csv_reader:
      # #if line_count == 0:
      # #  line_count = line_count + 1 #다음의 경우 헤더있음. 네이버는 이거 주석처리!!
      # if line_count > 10:
      #   break;
        
      # else:
        webtoon_idx = row[0]
        webtoon_name = row[1]
        webtoon_writer = row[2]
        webtoon_score = row[3]
        webtoon_agelimit = row[5]

        if len(webtoon_writer.split(' /')) > 1: 
          webtoon_writer = webtoon_writer.split(' /')[0]
          #print(webtoon_writer)
        
        my_writer.writerow([webtoon_idx, webtoon_name, webtoon_score, webtoon_writer, webtoon_agelimit])
        line_count = line_count +1
```

## 2. 장르 / 키워드와 합쳐서 최종 csv (naver_complete.csv)

* 아직 키워드를 좀더 가공해서 쓸지 어떻게 할지는 미완!

```
import pandas as pd
init_data = pd.read_csv('./crawling_result_naver_clear.csv')

genre_data = pd.read_csv('./crawling_genre_tagging_naver_v2.csv')

keyword_data = pd.read_csv('./storykeywords_result_naver.csv')

init_data['g1'] = genre_data['episode']
init_data['g2'] = genre_data['story']
init_data['g3'] = genre_data['daily_life']
init_data['g4'] = genre_data['comic']
init_data['g5'] = genre_data['fantasy']
init_data['g6'] = genre_data['action']
init_data['g7'] = genre_data['drama']
init_data['g8'] = genre_data['romance']
init_data['g9'] = genre_data['thriller']
init_data['g10'] = genre_data['period_drama']
init_data['g11'] = genre_data['sports']
init_data['g12'] = genre_data['campus_life']
init_data['keywords'] = keyword_data['스토리 키워드']

init_data.to_csv('./naver_complete.csv',index=False)
init_data.head()
```

# 다음 csv 정리

제목 / 작가 / 평점 / 연령정보(성인orNot) / 장르(g1 - g12) / 키워드  형태로

## 1. 기존의 크롤링 csv에서 clear형태로가공 (crawling_result_daum_clear.csv)

```
with open('crawling_result_daum.csv', encoding="utf-8") as csv_file:
  with open('crawling_result_daum_clear.csv',mode='w',newline='',encoding='utf-8') as my_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    my_writer = csv.writer(my_file,delimiter=',')
    my_writer.writerow(['idx','name','score','writer','age_limit'])   #,'g1','g2','g3','g4','g5','g6','g7','g8','g9','g10','g11','g12', 'keywords'])

    line_count = 0
    for row in csv_reader:
      if line_count == 0:
       line_count = line_count + 1 #다음의 경우 헤더있음
      # elif line_count > 10:
      #    break;
        
      else:
        idx_int = int(row[0])
        webtoon_idx = idx_int + 1182
        webtoon_name = row[1]
        webtoon_writer = row[2]
        webtoon_score = row[3]
        webtoon_agelimit = row[4]

        if len(webtoon_writer.split("', '")) > 1: 
          webtoon_writer = webtoon_writer.split("', '")[0][2:]
          
        webtoon_score = webtoon_score[0:4]
        my_writer.writerow([webtoon_idx, webtoon_name, webtoon_score, webtoon_writer, webtoon_agelimit])
        line_count = line_count +1
```
