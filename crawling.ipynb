```python
#############################################################
#                     Naver Webtton Crwaling                #
#############################################################

import requests
import csv
from bs4 import BeautifulSoup

f = open('crawling_result.csv', 'w', newline='', encoding='utf-8')
wr = csv.writer(f)

url_episode = 'https://comic.naver.com/webtoon/genre.nhn'
url_omnibus = 'https://comic.naver.com/webtoon/genre.nhn?genre=omnibus'
url_story = 'https://comic.naver.com/webtoon/genre.nhn?genre=story'

url_list = [url_episode, url_omnibus, url_story]
title_list = []
for i in url_list:
    response = requests.get(i)
    soup = BeautifulSoup(response.text, 'html.parser')
    temp = soup.find_all('dl')
    for j in temp:
        title_list.append(j)

cnt=0
for a in title_list:
    cnt+=1
    print(cnt, "번째 웹툰 크롤링 중...")
    webtoon_info = [cnt]
    webtoon_info.append(a.find('a').get_text())
    print("제목 : ", a.find('a').get_text())
    webtoon_info.append(a.find('dd').get_text()[1:len(a.find('dd').get_text())-1])
    print("작가 : ", a.find('dd').get_text()[1:len(a.find('dd').get_text())-1])
    webtoon_info.append(a.find('strong').get_text())
    print("평점 : ", a.find('strong').get_text())
    webtoon_url =  "https://comic.naver.com" + a.find('a').get('href')
    response = requests.get(webtoon_url)
    soup = BeautifulSoup(response.text, 'html.parser')
    webtoon_info.append(soup.find('span', class_='genre').get_text())
    print("장르 : ", soup.find('span', class_='genre').get_text())
    try:
        age=soup.find('span', class_='age').get_text()
        if age not in "18세 이용가":
          age="-"
        print("연령 : ", age)
    except:
        age="-"
        print('연령 :', age)
    webtoon_info.append(age)

    description=soup.find('div',class_='detail')
    webtoon_info.append(description.find('p').get_text())
    print("한 줄 소개 : ",description.find('p').get_text())

    wr.writerow(webtoon_info)
    print()

print(cnt, "개 웹툰 크롤링 종료")
print("출력 파일 : crawling_result")
```

```python
#############################################################
#                     Naver Webtoon Crwaling                #
#############################################################

# Crawling last episode links of each webtoon

import requests
import csv
from bs4 import BeautifulSoup
import os.path

f = open('crawling_result.csv', 'w', newline='', encoding='utf-8')
wr = csv.writer(f)

url= 'https://comic.naver.com'
url_episode = 'https://comic.naver.com/webtoon/genre.nhn'
url_omnibus = 'https://comic.naver.com/webtoon/genre.nhn?genre=omnibus'
url_story = 'https://comic.naver.com/webtoon/genre.nhn?genre=story'

url_list = [url_episode, url_omnibus, url_story]

link_list = []
idx_title_list = []
title_idx_list = []
cnt = 0

for i in url_list:
    response = requests.get(i)
    soup = BeautifulSoup(response.text, 'html.parser')
    webtoons = soup.find_all('dt')
    for webtoon in webtoons:
        link=webtoon.find('a').get('href')
        response2 = requests.get(url+link)
        soup2 = BeautifulSoup(response2.text, 'html.parser')
        title = soup2.find('img').get('alt')
        #print(title)
        res = soup2.select('td>a')[2] 
        res = res.get('href')
        #print(res)
        link_list.append( str("\""+title+"\" : \""+url+res+"\"") )
        idx_title_list.append( str("\""+str(cnt)+"\" : \""+title+"\"") )
        title_idx_list.append( str("\""+title+"\" : \""+str(cnt)+"\"") )
        cnt = cnt+1
    print("done crawling from ",i)
print("done :",cnt)

with open("naver_links.txt", 'w') as handle:
    for row in link_list:
            s = "".join(map(str, row))
            handle.write(s+',\n')
with open("naver_idx_to_class.txt", 'w') as handle:
    for row in idx_title_list:
            s = "".join(map(str, row))
            handle.write(s+',\n')
with open("naver_class_to_idx.txt", 'w') as handle:
    for row in title_idx_list:
            s = "".join(map(str, row))
            handle.write(s+',\n')
    
```
